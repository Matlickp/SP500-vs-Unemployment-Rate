```python
import pandas as pd
```

# NBER.org Mortality data per year per month

```python
# Data was downloaded from https://www.nber.org/data/vital-statistics-mortality-data-multiple-cause-of-death.html
# Downloaded files are not included because of the size of the individual files.
```

```python
# Read individual CSV files and create a list of dataframes.

#setting filepath
filepath = "Data/NBER_HF_Data/"
HD_Deaths = []

for yr in range (2000, 2018):
    filename = "mort"+ str(yr) +".csv"
    
    #read csv file for each year
    df = pd.read_csv(filepath + filename, usecols=['year','monthdth', 'ucod'])
    
    #filter rows with ICD10 code = 'I50.*' (Heart failure)
    hf_df = df.loc[ df['ucod'].str.startswith('I50', na=False)]
    
    #Group by Year and month of death
    hf_grp_by_yr = hf_df.groupby(['year','monthdth'])
    
    #Count the number of deaths per month & year
    hf_deaths = hf_grp_by_yr['ucod'].count()
    hf_deaths = hf_deaths.reset_index()
    
    print (hf_deaths)
    
    print ("Type is ", type(hf_deaths))

    
    #Append to the list
    HD_Deaths.append(hf_deaths)
    
```

```python

#Convert list of Dataframes to a single dataframe
Combined_Df = pd.concat(HD_Deaths)

#create a date colummn with month and Year
Combined_Df['date'] = '1/'+ Combined_Df['monthdth'].astype(str) + '/' + Combined_Df['year'].astype(str)
Combined_Df.head()
```

```python
#write to csv file
Combined_Df.to_csv('Data/Combined_HF_2000_2017.csv', index=False)
```
