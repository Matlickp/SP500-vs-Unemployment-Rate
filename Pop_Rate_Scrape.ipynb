```python
spy_data = pd.read_csv('spy.csv', parse_dates=True)
spy_data.head()
```

```python
# Create a column that shows the 100 Day Moving Average
spy_data['100MA'] = spy_data['Adj Close'].rolling(window=100, min_periods=0).mean()

# Drop unneccessary columns
spy_data = spy_data.drop(["High","Low","Open","Close"], axis=1)
spy_data.head()
```

```python
spy_data.dtypes
```

```python
spy_data.rename(columns={"Volume":"SPY Volume", "Adj Close":"SPY AdjClose", "100MA":"SPY 100MA"}, inplace=True)
spy_data.head()
```

```python
vix_data = pd.read_csv('vix.csv', parse_dates=True)
vix_data.head()
```

```python
vix_data.rename(columns={"Volume":"VIX Volume", "Adj Close":"VIX AdjClose", "100MA":"VIX 100MA"}, inplace=True)
vix_data.head()
```

```python
vix_data.dtypes
```

```python
merge_data = pd.merge(vix_data, spy_data, how = "outer", on="Date")
merge_data = merge_data.reset_index()
merge_data.head()
```

```python
merge_data.dtypes
```

```python
merge_data = pd.merge(merge_data, em_data, on = "Date", how = "outer")
merge_data.head()
```

```python
merge_data = pd.merge(merge_data, pop_rate_year, on = "Date", how = "outer")
merge_data.head()
```
